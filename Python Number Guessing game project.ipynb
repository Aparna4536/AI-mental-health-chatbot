Hill Cipher

```python
import numpy as np

# Function to find modular inverse
def mod_inverse(a, m):
    for x in range(1, m):
        if (a * x) % m == 1:
            return x
    return None

# Function to process text
def process_text(text):
    text = text.upper().replace(" ", "")
    if len(text) % 2 != 0:
        text += 'X'  # padding
    return text

# Encryption
def hill_encrypt(plaintext, key_matrix):
    plaintext = process_text(plaintext)
    cipher_text = ""

    for i in range(0, len(plaintext), 2):
        pair = plaintext[i:i+2]
        vector = np.array([[ord(pair[0]) - 65],
                           [ord(pair[1]) - 65]])

        result = np.dot(key_matrix, vector) % 26
        cipher_text += chr(result[0][0] + 65)
        cipher_text += chr(result[1][0] + 65)

    return cipher_text

# Decryption
def hill_decrypt(ciphertext, key_matrix):
    det = int(np.linalg.det(key_matrix))
    det = det % 26
    det_inv = mod_inverse(det, 26)

    if det_inv is None:
        raise ValueError("Key matrix is not invertible")

    adjugate = np.round(det * np.linalg.inv(key_matrix)).astype(int) % 26
    inverse_matrix = (det_inv * adjugate) % 26

    plain_text = ""

    for i in range(0, len(ciphertext), 2):
        pair = ciphertext[i:i+2]
        vector = np.array([[ord(pair[0]) - 65],
                           [ord(pair[1]) - 65]])

        result = np.dot(inverse_matrix, vector) % 26
        plain_text += chr(result[0][0] + 65)
        plain_text += chr(result[1][0] + 65)

    return plain_text

# Example Usage
key = np.array([[3, 3],
                [2, 5]])

plaintext = "HELLO"

cipher = hill_encrypt(plaintext, key)
print("Encrypted:", cipher)

decrypted = hill_decrypt(cipher, key)
print("Decrypted:", decrypted)
```

2) Validate the Key Matrix(Check if Invertible Modulo 26)

```python
# Function to calculate determinant of 2x2 matrix
def determinant(matrix):
    return matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]

# Function to find modular inverse
def mod_inverse(a, m):
    a = a % m
    for x in range(1, m):
        if (a * x) % m == 1:
            return x
    return None

det = determinant(key) % 26
inv = mod_inverse(det, 26)

if inv is None:
    print("Key Matrix is NOT invertible modulo 26")
else:
    print("Key Matrix is invertible modulo 26")
```

4) Encrypt the plaintext using matrix multiplication and modulo operations.

```python
# Example key matrix (2x2)
key = [[3, 3],
       [2, 5]]

# Example plaintext numbers
plain_numbers = [7, 4, 11, 11, 14]   # Example: HELLO

cipher_numbers = []

for i in range(0, len(plain_numbers), 2):

    block = plain_numbers[i:i+2]

    # Padding if block has only one element
    if len(block) < 2:
        block.append(23)   # X = 23

    c1 = (key[0][0] * block[0] + key[0][1] * block[1]) % 26
    c2 = (key[1][0] * block[0] + key[1][1] * block[1]) % 26

    cipher_numbers.append(c1)
    cipher_numbers.append(c2)

print("Cipher numbers:", cipher_numbers)
```

5) Compute the inverse of the key matrix for decryption.

```python
# Example key matrix
key = [[3, 3],
       [2, 5]]

# Function to calculate determinant of 2x2 matrix
def determinant(matrix):
    return matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]

# Function to find modular inverse
def mod_inverse(a, m):
    a = a % m
    for x in range(1, m):
        if (a * x) % m == 1:
            return x
    return None

det = determinant(key) % 26
inv_det = mod_inverse(det, 26)

# Check if inverse exists
if inv_det is None:
    print("Key matrix is not invertible modulo 26")
else:
    inverse_key = [
        [(key[1][1] * inv_det) % 26, (-key[0][1] * inv_det) % 26],
        [(-key[1][0] * inv_det) % 26, (key[0][0] * inv_det) % 26]
    ]

    print("Inverse Key Matrix:")
    for row in inverse_key:
        print(row)
```

6) Decrypt the ciphertext using the inverse key matrix to recover the original plaintext.

```python
# Example ciphertext
ciphertext = "HIAT"

# Example inverse key matrix
inverse_key = [[15, 17],
               [20, 9]]

# Convert ciphertext to numbers
cipher_numbers = [ord(c) - ord('A') for c in ciphertext]

plain_numbers = []

for i in range(0, len(cipher_numbers), 2):

    block = cipher_numbers[i:i+2]

    # Padding safety if block has only one number
    if len(block) < 2:
        block.append(23)

    p1 = (inverse_key[0][0] * block[0] + inverse_key[0][1] * block[1]) % 26
    p2 = (inverse_key[1][0] * block[0] + inverse_key[1][1] * block[1]) % 26

    plain_numbers.append(p1)
    plain_numbers.append(p2)

decrypted_text = ""

for num in plain_numbers:
    decrypted_text += chr(num + ord('A'))

print("Decrypted Text:", decrypted_text)
```

8. Largest sum contiguous subarray (Kadane’s Algorithm)

```python
arr = [-2,1,-3,4,-1,2,1,-5,4]

max_sum = arr[0]
current_sum = arr[0]

for i in range(1, len(arr)):
    current_sum = max(arr[i], current_sum + arr[i])
    max_sum = max(max_sum, current_sum)

print(max_sum)
```

Python Number Guessing game project

```python
import random

def guess_the_number():
    # Generate random number between 1 and 100
    number = random.randint(1, 100)

    print("Welcome to Guess the Number Game!")
    print("Guess a number between 1 and 100\n")

    while True:
        try:
            # Take user input
            guess = int(input("Enter your guess: "))

            # Calculate difference
            difference = abs(guess - number)

            # Conditions
            if difference == 0:
                print("Correct! You guessed the number.")
                break
            elif difference <= 10:
                print("Hot")
            else:
                print("Cold")

        except ValueError:
            print("Invalid input! Please enter a number.")

# Run the game
guess_the_number()
```
